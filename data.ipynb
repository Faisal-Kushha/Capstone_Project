```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('./DA_DS_Data.csv')
print(df)
df.info()
```

```python
description = df.iloc[:,(4)]

skills = []

requirment = ["SQL", "Python","Excel", "Tableau", "dashboard", "data modeling","data mining", "data pipline",
 "data warehouse","data analysis","statistic", "report","data visualization","PowerBI", "R ","AWS","Looker",
 "Hadoop", "Spark", "BigQuery","data wrangling", "Big Data", "Statistics", "Data Mining", "Machine Learning", "NLP",
 "Microsoft SQL", "Oracle", "Azure","ETL", "GCP", "ML ", "Deep Learning", " Artificial Intelligence", "AI ", 
 "data engineer", "Neural Networks", "algorithm", "data integration", "mathematics", "Model Planning", 
 "data extraction", "data cleaning", "data querying", "NoSQL"]

for column in description:
    x = ''
    for i in requirment:
        if i in column:
           x = x + i + ", "
    skills.append(x)
```

```python
df["Skills"]=pd.Series(skills)
df
```

```python
location = df.iloc[:,(1)]
state = []

california = ["CA", "California", "california", "CALIFORNIA"]
new_York = ["NY", "new york", "New York", "NEW YORK"]
new_Jersey = ["NJ", "new jersey", "New Jersey", "NEW JERSEY"]
pennsylvania = ["PA", "Pennsylvania", "pennsylvania", "PENNSYLVANIA"]
texas = ["TX", "Texas", "texas", "TEXAS"]
virginia = ["VA", "Virginia", "virginia", "VIRGINIA"]
washington = ["WA", "Washington", "washington", "WASHINGTON"]
georgia = ["GA", "Georgia", "georgia", "GeORGIA"]
maryland = ["MD", "Maryland", "maryland", "MARYLAND"]
illinois = ["IL", "Illinois", "illinois", "ILLINOIS"]

for column in location:
    x = ''
    for i in california:
        if i in column:
           x = x + "CA" 
    for i in new_York:
        if i in column:
           x = x + "NY" 
    for i in new_Jersey:
        if i in column:
           x = x + "NJ" 
    for i in pennsylvania:
        if i in column:
           x = x + "PA" 
    for i in texas:
        if i in column:
           x = x + "TX" 
    for i in virginia:
        if i in column:
           x = x + "VA" 
    for i in washington:
        if i in column:
           x = x + "WA"
    for i in georgia:
        if i in column:
           x = x + "GA" 
    for i in maryland:
        if i in column:
           x = x + "MD" 
    for i in illinois:
        if i in column:
           x = x + "IL" 
    state.append(x)

```

```python
df["State"]=pd.Series(state)
df
```

```python
location = df.iloc[:,(1)]

environment = []

for column in location:
    x = ''
    if ("Hybrid " or "Hybrid remote ") in column:
        x = x + "Hybrid"
    elif "Remote " in column:
        x = x + "Remote" 
    else:
        x = x + "Office"
    environment.append(x)
```

```python
df["Workplace"]=pd.Series(environment)
df
```

```python
df.drop('location', axis=1, inplace=True)
df.drop('desc', axis=1, inplace=True)
df
```

```python
df.rename(columns = {'title':'Job Title', 'company':'Company','rating':'Rating','job_tybe':'Job Type',
'job':'Category','final_salary':'Avg Annual Salary'}, inplace = True)
df
```
